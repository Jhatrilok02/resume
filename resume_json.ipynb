```python
import json

def parse_resume(content):
    resume = {
        "Personal Information": {
            "Name": "",
            "Contact Information": {
                "Email": "",
                "Phone": ""
            }
        },
        "Education": [],
        "Experience": [],
        "Projects": [],
        "Skills": [],
        "Certifications": []
    }

    lines = content.strip().split("\n")

    # Parse each line based on expected format
    for line in lines:
        line = line.strip()
        if line.startswith("Name:"):
            resume["Personal Information"]["Name"] = line.replace("Name:", "").strip()
        elif line.startswith("Email:"):
            resume["Personal Information"]["Contact Information"]["Email"] = line.replace("Email:", "").strip()
        elif line.startswith("Phone:"):
            resume["Personal Information"]["Contact Information"]["Phone"] = line.replace("Phone:", "").strip()
        elif line.startswith("Education:"):
            education_line = line.replace("Education:", "").strip()
            education_parts = education_line.split(",")
            if len(education_parts) == 3:
                degree, university, graduation_year = [x.strip() for x in education_parts]
                resume["Education"].append({
                    "Degree": degree,
                    "University": university,
                    "Graduation Year": graduation_year
                })
        elif line.startswith("Experience:"):
            experience_line = line.replace("Experience:", "").strip()
            experience_parts = experience_line.split(",")
            if len(experience_parts) == 4:
                job_title, company, duration, responsibilities = [x.strip() for x in experience_parts]
                resume["Experience"].append({
                    "Job Title": job_title,
                    "Company": company,
                    "Duration": duration,
                    "Responsibilities": responsibilities
                })
        elif line.startswith("Projects:"):
            projects_line = line.replace("Projects:", "").strip()
            for project in projects_line.split(";"):
                project_parts = project.strip().split(",")
                if len(project_parts) >= 3:
                    title = project_parts[0].strip()
                    description = project_parts[1].strip()
                    technologies = ", ".join(project_parts[2:]).strip()  # Join remaining parts as technologies
                    resume["Projects"].append({
                        "Title": title,
                        "Description": description,
                        "Technologies Used": technologies
                    })
        elif line.startswith("Skills:"):
            resume["Skills"] = [skill.strip() for skill in line.replace("Skills:", "").strip().split(",")]
        elif line.startswith("Certifications:"):
            resume["Certifications"] = [certification.strip() for certification in line.replace("Certifications:", "").strip().split(",")]

    return resume

def main():
    resume_content = """
    Name: Trilok Nath Jha
    Email: trilok@example.com
    Phone: +91-1234567890
    Education: B.Tech in Computer Science, Lovely Professional University, 2025
    Experience: Machine Learning Intern at Zebo.ai, May 2022 - July 2022, Worked on model preprocessing, training/testing, and predictive modeling
    Projects: Customer Churn Prediction, Analyzed customer behavior patterns to predict churn, Python; Stock Market Prediction, Predicted stock prices using regression models, Python; Iris Classification, Classified iris species using SVM, Python
    Skills: Python, Java, SQL, HTML, CSS, C#, Machine Learning
    Certifications: Machine Learning with Python, Solving Competitive Problems using Data Structures, Desktop Application Development using JAVA, Natural Language Processing
    """

    resume_json = parse_resume(resume_content)
    print(json.dumps(resume_json, indent=4))

if __name__ == "__main__":
    main()
```
